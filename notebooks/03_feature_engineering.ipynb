```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/clean_instagram_data.csv")
```

```python
df["engagement_score"] = (
    df["likes"] +
    2 * df["comments"] +
    3 * df["shares"] +
    2 * df["saves"]
)
```

```python
df["normalized_engagement"] = df["engagement_score"] / df["follower_count"]
```

```python
df["is_weekend"] = df["day_of_week"].isin(["saturday", "sunday"]).astype(int)
```

```python
df["is_weekend"] = df["day_of_week"].isin(["saturday", "sunday"]).astype(int)
```

```python
df["has_cta"] = df["has_call_to_action"].astype(int)
```

```python
df["caption_density"] = df["caption_length"] / (df["hashtags_count"] + 1)
```

```python
viral_threshold = df["normalized_engagement"].quantile(0.90)
df["is_viral"] = (df["normalized_engagement"] >= viral_threshold).astype(int)
```

```python
df["is_viral"].value_counts(normalize=True)
```

```python
final_features = [
    "account_type",
    "media_type",
    "content_category",
    "traffic_source",
    "follower_count",
    "hashtags_count",
    "caption_length",
    "has_cta",
    "is_weekend",
    "post_hour",
    "likes",
    "comments",
    "shares",
    "saves"
]

df_model = df[final_features + ["normalized_engagement", "is_viral"]]
```

```python
df_model.to_csv("../data/final_instagram_model_data.csv", index=False)
```
